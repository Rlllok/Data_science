```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_datasets as tfds

tfds.disable_progress_bar()
```

```python
(raw_train, raw_validation, raw_test), metadata = tfds.load(
    'cats_vs_dogs',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    with_info=True,
    as_supervised=True,
)
```

```python
IMG_SIZE = 160
```

```python
get_label_name = metadata.features['label'].int2str # function that returns label
```

```python
def format_image(image, label):
    image = tf.cast(image, tf.float32)
    image = (image / 127.5) - 1
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    return image, label
```

```python
train = raw_train.map(format_image)
validation = raw_validation.map(format_image)
test = raw_test.map(format_image)
```

```python
BATCH_SIZE = 32
SHUFFLE_BUFFER_SIZE = 1000

train_batches = train.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
validation_batches = validation.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
test_batches = test.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
```

```python
# PICKING A PRETRAINED MODEL
IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)

base_model = tf.keras.applications.MobileNetV2(input_shape=IMG_SHAPE,
                                               include_top=False,
                                               weights='imagenet')
```

```python
base_model.summary()
```

```python
# FREEZING THE MODEL

base_model.trainable = False
base_model.summary()
```

```python
# ADDING CLASSIFIER
global_avarage_layer = tf.keras.layers.GlobalAveragePooling2D()
prediction_layer = tf.keras.layers.Dense(1)

model = tf.keras.Sequential([
    base_model,
    global_avarage_layer,
    prediction_layer
])
```

```python
model.summary()
```

```python
# TRAINING THE MODEL

base_learning_rate = 0.0001
model.compile(optimizer=tf.keras.optimizers.RMSprop(lr=base_learning_rate),
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])

initial_epochs = 3
validation_steps = 20

# loss, accuracy = model.evaluate(validation_batches, steps=validation_steps)
```

```python
history = model.fit(train_batches,
                    epochs=initial_epochs,
                    validation_data=validation_batches)
```

```python
acc = history.history['accuracy']
print(acc)
```

```python
model.save("dogs_vs_cats.h5")  # we can save the model and reload it at anytime in the future
new_model = tf.keras.models.load_model('dogs_vs_cats.h5')
```
