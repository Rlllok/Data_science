```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('Solarize_Light2')
```

```python
train_df = pd.read_csv('train.csv')
train_df['source'] = 'train'
test_df = pd.read_csv('test.csv')
test_df['source'] = 'test'
```

```python
data = pd.concat([train_df, test_df], ignore_index=True)
```

```python
def get_null_features(data):
    features = data.columns
    null_features = []

    for feature in features:
        if data[feature].isnull().sum() != 0:
            null_features.append(feature)
    return null_features
```

```python
data.head()
```

```python
# Fill 'MSZoning' NaN values with RL (mode)
data['MSZoning'] = data['MSZoning'].fillna('RL')

# Fill 'LotFrontage' NaN values with mean
data['LotFrontage'] = data['LotFrontage'].fillna(data['LotFrontage'].mean())

# Fill 'Alley' NaN values with None (no alley access)
data['Alley'] = data['Alley'].fillna('None')

# Fill 'Utilities' NaN with AllPub (mode)
data['Utilities'] = data['Utilities'].fillna('AllPub')

# Fill 'Exterior1st' NaN with VinylSd (mode)
data['Exterior1st'] = data['Exterior1st'].fillna('VinylSd')

# Fill 'Exterior2nd' NaN with VinylSd (mode)
data['Exterior2nd'] = data['Exterior2nd'].fillna('VinylSd')

# Fill 'MasVnrType' NaN with None (mode)
data['MasVnrType'] = data['MasVnrType'].fillna('None')

# Fill 'MasVnrArea' NaN with 0 (mode)
data['MasVnrArea'] = data['MasVnrArea'].fillna(0)

# Fill 'BsntQual' NaN with None (no basement)
data['BsmtQual'] = data['BsmtQual'].fillna('None')

# Fill 'BsmtCond' NaN with None (no basement)
data['BsmtCond'] = data['BsmtCond'].fillna('None')

# Fill 'BsmtExposure' NaN with None (no basement)
data['BsmtExposure'] = data['BsmtExposure'].fillna('None')

# Fill 'BsmtFinType1' NaN with None (no basement)
data['BsmtFinType1'] = data['BsmtFinType1'].fillna('None')

# Fill 'BsmtFinSF1' NaN with 0 (mode)
data['BsmtFinSF1'] = data['BsmtFinSF1'].fillna(0)

# Fill 'BsmtFinType2' NaN with None (no basement)
data['BsmtFinType2'] = data['BsmtFinType2'].fillna('None')

# Fill 'BsmtFinSF2' NaN with 0 (mode)
data['BsmtFinSF2'] = data['BsmtFinSF2'].fillna(0)

# Fill 'BsmtUnfSF' NaN with 0 (mode)
data['BsmtUnfSF'] = data['BsmtUnfSF'].fillna(0)

# Fill 'TotalBsmtSF' NaN with 0 (mode)
data['TotalBsmtSF'] = data['TotalBsmtSF'].fillna(0)

# Fill 'Electrical' NaN with SBrkr (mode)
data['Electrical'] = data['Electrical'].fillna('SBrkr')

# Fill 'BsmtFullBath' NaN with SBrkr (mode)
data['BsmtFullBath'] = data['BsmtFullBath'].fillna(0)

# Fill 'BsmtHalfBath' NaN with 0 (mode)
data['BsmtHalfBath'] = data['BsmtHalfBath'].fillna(0)

# Fill 'KitchenQual' NaN with TA (mode)
data['KitchenQual'] = data['KitchenQual'].fillna('TA')

# Fill 'Functional' NaN with Typ (mode)
data['Functional'] = data['Functional'].fillna('Typ')

# Fill 'FireplaceQu' NaN with None (no fireplace)
data['FireplaceQu'] = data['FireplaceQu'].fillna('None')

# Fill 'GarageType' NaN with None (no garage)
data['GarageType'] = data['GarageType'].fillna('None')

# Fill 'GarageCars' NaN with 2 (mode)
data['GarageCars'] = data['GarageCars'].fillna(2)

# Fill 'GarageYrBlt' NaN with 2005 (mode)
data['GarageYrBlt'] = data['GarageYrBlt'].fillna(2005)

# Fill 'GarageFinish' NaN with None (no garage)
data['GarageFinish'] = data['GarageFinish'].fillna('None')

# Fill 'GarageArea' NaN with 0 (mode)
data['GarageArea'] = data['GarageArea'].fillna(0)

# Fill 'GarageQual' NaN with None (no garage)
data['GarageQual'] = data['GarageQual'].fillna('None')

# Fill 'GarageCond' NaN with None (no garage)
data['GarageCond'] = data['GarageCond'].fillna('None')

# Fill 'PoolQC' NaN with None (no pool)
data['PoolQC'] = data['PoolQC'].fillna('None')

# Fill 'Fence' NaN with None (no fence)
data['Fence'] = data['Fence'].fillna('None')

# Fill 'MiscFeature' NaN with None
data['MiscFeature'] = data['MiscFeature'].fillna('None')

# Fill 'SaleType' NaN with WD (mode)
data['SaleType'] = data['SaleType'].fillna('WD')
```

```python
features = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt', 'SalePrice']
```

```python
data['SalePrice'] = np.log(data['SalePrice'])
```

```python
data['GrLivArea'] = np.log(data['GrLivArea'])
```

```python
data['Bsmt'] = data['TotalBsmtSF']
data.loc[data['TotalBsmtSF']>0, 'Bsmt'] = 1
data.loc[data['Bsmt']==1, 'TotalBsmtSF'] = np.log(data['TotalBsmtSF'])
```

```python
train = data.loc[data['source']=='train', ['Id']+features]
test = data.loc[data['source']=='test', ['Id']+features]
test = test.drop('SalePrice', axis=1)
```

```python
from sklearn import metrics

def print_evaluate(y_true, y_pred):
    mae = metrics.mean_absolute_error(y_true, y_pred)
    mse = metrics.mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    rmsle = metrics.mean_squared_log_error(y_true, y_pred)
    r2 = metrics.r2_score(y_true, y_pred)
    
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2', r2)
```

```python
from sklearn.model_selection import train_test_split

X = train.drop(['Id', 'SalePrice'], axis=1)
y = train['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
xgb.fit(X, y)
y_pred = xgb.predict(test.drop('Id', axis=1))
y_pred = np.exp(y_pred)
```

```python
submission = pd.DataFrame(columns=['Id', 'SalePrice'])
submission['Id'] = test['Id']
submission['SalePrice'] = y_pred
submission.to_csv('best_submission.csv', index=False)
```
